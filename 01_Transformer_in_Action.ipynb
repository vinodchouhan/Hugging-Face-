```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life", "I hate this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education","politics","business"],
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=3)
```

```python
from transformers import pipeline
ner = pipeline("ner", grouped_entities=True)
ner("My name is Vinod and I work at EXL in Hyderabad.")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
                  question="Where do I work?",
                  context="My name is Vinod and I work at EXL in Hyderabad."
                  )
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    Modern India is a dynamic and diverse nation rapidly evolving into a global powerhouse. Since gaining independence in 1947, India has transformed from a primarily agrarian economy into one of the world’s largest and fastest-growing economies. It is now a hub for technology, innovation, and services, with cities like Bengaluru, Hyderabad, and Pune leading the digital revolution.
India is also the world’s most populous country, home to over 1.4 billion people. It thrives on its cultural richness, democratic values, and entrepreneurial spirit. Modern India balances ancient traditions with modern ambitions, embracing advancements in space research, artificial intelligence, renewable energy, and digital infrastructure.
Challenges such as inequality, climate change, and urbanization persist, but India’s young population, policy reforms, and global collaborations position it as a key player in shaping the 21st century.
From startups to space missions, modern India is a story of resilience, growth, and global aspiration.
    """
)
```
